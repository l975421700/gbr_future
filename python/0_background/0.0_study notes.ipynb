# WRF

## Compiling

### WRF on gadi: https://github.com/coecms/WRF

#### Install

    $ git ls-remote --heads https://github.com/coecms/WRF.git
    $ git clone -b V4.5.2 --recursive https://github.com/coecms/WRF.git

#### Build WRF

    $ cd WRF/WRF/
    $ ./run_compile -h
    $ ./run_compile
    $ ls -l main/*.exe


#### Build WPS

    $ cd WRF/WPS/
    $ ./run_compile -h
    $ ./run_compile
    $ ls -l *.exe
    $ ls -l */src/*.exe
